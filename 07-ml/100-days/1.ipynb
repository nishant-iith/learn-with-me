```python
class LR:
    # constructor
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, x_train, y_train):
        xbar = x_train.mean()
        ybar = y_train.mean()

        m = ((x_train - xbar) * (y_train - ybar)).sum() / ((x_train - xbar) ** 2).sum()
        b = ybar - m * xbar

        self.m = m
        self.b = b

    def predict(self, x_test):
        return self.m * x_test + self.b
```

```python
# Example dataset
import numpy as np
import pandas as pd
```

```python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Exam_Score': [50, 55, 65, 70, 75]
}

df = pd.DataFrame(data)
```

```python
x_train = df['Hours_Studied']
y_train = df['Exam_Score']
```

```python
lr = LR()
lr.fit(x_train, y_train)
```
